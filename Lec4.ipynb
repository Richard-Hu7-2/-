#### 如何计算 cret 12

```python
stock['Mretnd'].iloc[:30]
```

```python
tp = (np.exp(stock.groupby(['Stkcd'])['Mretnd'].rolling(12).sum())-1).reset_index(level=0)
```

```python
stock['cret12'] = tp['Mretnd']  # 危险！！！

## 列对列赋值  —————— index 匹配
```

```python
stock['cret12_lag'] = stock.groupby(['Stkcd'])['cret12'].shift(1)
```

#### sorting --- 分组  pd.qcut

```python
stock['rank'] = stock.groupby(['Trdmnt'])['cret12_lag'].rank(ascending=True)
```

```python
stock['cret12_decile'] = stock.dropna(subset=['cret12_lag']).groupby(['Trdmnt'])['cret12_lag'].transform(lambda x:pd.qcut(x,5,labels=False))
```

### 计算portfolio 收益

```python
stock
```

```python
tp = stock.groupby(['Trdmnt','cret12_decile'])['Mretwd'].mean().reset_index()
```

```python
# 
tp = stock.groupby(['Trdmnt','cret12_decile']).apply(lambda x:(x['Msmvttl']*x['Mretwd']).sum() / x['Msmvttl'].sum()).reset_index()
```

```python
tp.groupby(['Trdmnt']).apply(lambda x:x.loc[x['cret12_decile']==4,'Mretwd'].iloc[0] - x.loc[x['cret12_decile']==0,'Mretwd'].iloc[0])
```

```python
# pivot  将长形数据 变成 宽型数据

ttp = tp.pivot(index='Trdmnt',columns='cret12_decile',values='Mretwd')
ttp
```
