```python
res = requests.get('https://news.ruc.edu.cn/archives/category/important_news',headers=headers)
contents = res.content.decode('utf-8')
from bs4 import BeautifulSoup

di = {}
counter = 0

while True:

    soup = BeautifulSoup(contents,'html.parser')
    res = soup.select('#content > div > div.content_col_2 > div.content_col_2_list > ul > li')
    
    
    for i in res:
        di[counter] = (i.a.text,i.a['href'],i.span.text)
        counter += 1

    
    url = soup.select('#content > div > div.content_col_2 > div.content_col_2_nav > div.content_col_2_nav_alignright > a')[0]['href'] # next page
    if url == None:
        break
    res = requests.get(url,headers=headers)
    contents = res.content.decode('utf-8')

    counter += 1
    print(f'post {counter} over!')
    if counter > 300:
        break
```

```python
url = 'https://news.ruc.edu.cn/archives/category/important_news'
#url = 'http://guba.eastmoney.com/news,600506,1298441713.html'

browser = webdriver.Chrome(executable_path=r"C:\Program Files\Google\Chrome\Application\chromedriver.exe") # you need to download a chromedriver adapted your version

browser.get(url)
print(browser.page_source) # equals to res.contents in requests
```

### True mimicking click by PyAutoGUI

```python
import pyautogui as pg
import time 
```

```python
p1 = (770,150)
p2 = (870,580)
p3_normal = (800,650)
p3_firstpage = (800,630)
p4 = (1170,580)
p5 = (1900,975)
```

```python
def drag2Nclick(point,dur=0.5): # drag cursor to point using dur s
    pg.dragTo(point[0],point[1],duration=dur)
    pg.click()
```

```python
time.sleep(5)

try:
    pli = [p1,p2,p3_firstpage,p4,p5]
    for x in pli:
        drag2Nclick(x)

    time.sleep(1)
    pli_normal = [p1,p2,p3_normal,p4,p5]
    while True:
        for x in pli_normal:
            drag2Nclick(x)
            
except KeyboardInterrupt:
    print('InterruptedError')
```
